```python
from elevenlabs import generate, play, set_api_key, stream
from google.colab import userdata

set_api_key(userdata.get('ElevenLabs_API_Key'))

audio = generate(
    text="Hi! I'm the world's most advanced text-to-speech system, made by elevenlabs.",
    voice="Rachel"
)

play(audio, notebook=True)
```

# Voice Streaming

Saving for later. https://mpv.io/ is required to be installed on the local system for streaming to run. Need to sign-up for Colab Pro

```python

def text_stream():
    yield "Hi there, I'm Eleven "
    yield "I'm a text to speech API "

audio_stream = generate(
    text=text_stream(),
    voice="Nicole",
    model="eleven_monolingual_v1",
    stream=True
)

stream(audio_stream)
```
